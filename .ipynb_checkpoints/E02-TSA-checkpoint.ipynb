# 02.1 Find the values of rho for which a series becomes non-stationary. rho between 0 and 1 ([0,1]) \n",
`X(t) = Rho * X(t-1) + Er(t)`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Comentario Prueba
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def ts(rho):
    ts = pd.Series(np.random.rand(200) - 0.5, index=pd.date_range("2020-01-01", periods=200))
    
    for i in range(1, ts.shape[0]):
        ts.iloc[i] = ts.iloc[i] + rho * ts.iloc[i-1]    
    return ts
```

```python
arr = pd.Series(np.random.rand(101))  
pv = pd.Series(np.random.rand(101))  
t = pd.Series(np.random.rand(101))  
arr[0] = 0.0
t[0] = 0.05
```

```python
for i in range(1, arr.shape[0]):
    arr[i] = arr[i-1] + 0.01
    t[i] = 0.05
```

```python
for i in range(0, pv.shape[0]):
    tsr = ts(rho=arr[i])
    result = adfuller(tsr)
    pv[i] = result[1]
```

```python
plt.plot(arr, pv, arr, t, 'r--')
plt.xlabel('Rho')
plt.ylabel('P-Value')
plt.suptitle('Dickey Fuller Test of Stationarity')
plt.show()
```

- El siguiente Dataframe representa los valores de Rho para los cuales la serie de datos no es estacionaria, con sus respectivos p-values de la prueba Dickey Fuller.

```python
df = pd.concat([arr, pv], axis=1)
df.columns = ['rho', 'pvalue']
df = df.loc[df['pvalue'] > 0.05]
df
```

# 02.2 Download and explore two series from google trends

```python
ts_data = pd.read_csv('Data/Time_Series_google_trends.csv', skiprows=1)
ts_data.columns = ['date', 'time_series']
ts_data.date = pd.to_datetime(ts_data.date)
ts_data.set_index('date', inplace=True)
ts_data.head()
```

```python
df_data = pd.read_csv('Data/Dickey_Fuller_google_trends.csv', skiprows=1)
df_data.columns = ['date', 'Dickey_Fuller']
df_data.date = pd.to_datetime(df_data.date)
df_data.set_index('date', inplace=True)
df_data.head()
```

```python
df = pd.concat([ts_data, df_data], axis=1)
df.head()
```

```python
df.plot(figsize=(50,15), linewidth=5, fontsize=40, color=('b', 'r'))
plt.xlabel('Year', fontsize=50)
plt.suptitle('Serie de datos - Google Trends', fontsize=40);
```

```python
df[['time_series', 'Dickey_Fuller']].rolling(12).mean().plot(figsize=(50,15), linewidth=5, fontsize=40, color=('b', 'r'))
plt.xlabel('Year', fontsize=50)
plt.suptitle('Tendencia de datos', fontsize=40);
```

```python
df.diff().plot(figsize=(50,15), linewidth=5, fontsize=40, color=('b', 'r'))
plt.xlabel('Year', fontsize=50)
plt.suptitle('Estacionalidad', fontsize=40);
```

```python
result = adfuller(df['time_series'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
result = adfuller(df['time_series'].diff().iloc[1:])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
result = adfuller(df['Dickey_Fuller'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
result = adfuller(df['Dickey_Fuller'].diff().iloc[1:])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

- Como se observa en los resultados de la prueba de Dickey Fuller las dos series diferenciadas y no diferenciadas son estacionarias.
